```python
# Librerías
# Para leer/guardar archivos en formato pickle
import pickle

# Para el manejo y análisis de estructuras de datos
import pandas as pd
```

```python
# Leer el archivo
df = pd.read_pickle("datos/1-bikes-eda.pkl")
df.head()
```

```python
df.dtypes
```

Realizamos "encoding" (codificación) a la columna categórica "season", ya que algunos algoritmos de machine learning no pueden procesar datos si no son numéricos.

Se le va a dar orden, por eso se va a utilizar el método map, puesto que nosotros establecemos dicho orden.

```python
# Definir el mapa a aplicar
mapa_season = {"spring": 3, "summer": 4, "autumn": 2, "winter": 1}
```

```python
# Aplicar el método map
df["season"] = df["season"].map(mapa_season)
```

```python
df.head()
```

```python
# Para comprobar que tiene los valores adecuados
df["season"].unique()
```

```python
# Guardar en pickle
df.to_pickle('datos/3-bikes-encoding-con-orden.pkl')
```
